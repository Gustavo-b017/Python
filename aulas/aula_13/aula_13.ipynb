problema 1 - soma da coluna 1 e 2, em uma nova coluna

```python
data = {
    'coluna 1': [10,20,30,40,50],
    'coluna 2': [5, 10, 15, 20, 25],
    'coluna 3': ["a", "b", "c", "d", "e"],
    'coluna 4': ['x', 'y', 'z', 'w', 'v']
}
colu = []

df = pd.DataFrame(data)


def soma_coluna(df):

    for _ in range(len(df)):
        df['soma'] = df['coluna 1'] + df['coluna 2']
        
        colu.append(df['soma'])

    return df

soma_coluna(df)
```

problema 2 - soma da coluna 1 e 2 se for a ou c; subitracao quando for b ou d; multiplicacao quando for e 

```python
import pandas as pd
 
data = {
    'coluna 1': [10, 20, 30, 40, 50],
    'coluna 2': [5, 10, 15, 20, 25],
    'coluna 3': ["a", "b", "c", "d", "e"],
    'coluna 4': ['x', 'y', 'z', 'w', 'v']
}
 
df = pd.DataFrame(data)
colu = []
 
def opera(df):
    for i in range(len(df)):
        if df['coluna 3'][i] in ['a', 'c']:
            a = df['coluna 1'][i] + df['coluna 2'][i]
            colu.append(a)
        elif df['coluna 3'][i] in ['b', 'd']:
            b = df['coluna 1'][i] - df['coluna 2'][i]
            colu.append(b)
        elif df['coluna 3'][i] == 'e':
            c = df['coluna 1'][i] * df['coluna 2'][i]
            colu.append(c)
        else:
            colu.append(None)
 
    df['resultados'] = colu
    return df
 
opera(df)
```
